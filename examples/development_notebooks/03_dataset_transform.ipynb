```python
from hydroml.data.dataset import get_transformer
from hydroml.utils import helpers as h

# loading up all the parameters
parameters =h.read_yaml(config.get_transform_parameters_path())

# taking the parameters for the target feature
parameters = parameters[config.target_features[0]] 

# get transformer
Transformer = get_transformer('norm')

# make the transformer
t = Transformer(parameters=parameters)

# print the original, transformed and inverse transformed values
print('original', ds.dataset['dynamic'].sel(dynamic_feature=config.target_features[0]).values[0,0:5], end='\n\n')
print('transformed', ds.y[0,0:5].flatten(), end='\n\n')
print('inverse_transformed', t.inverse_transform(ds.y[0,0:5]).flatten(), end='\n\n')

```
